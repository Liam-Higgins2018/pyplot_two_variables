# Two Varibale plots
 Adapted from this website: https://matplotlib.org/tutorials/introductory/pyplot.html

```python
import matplotlib.pyplot as plt
import numpy as np
```

$$ y = 2x $$

```python
x = np.linspace(0.0, 100.0, 11) 
y1 = 2.0 * x + 5.0
y2 = 1.0 * x - 5.0
y3 = -2.0 * x + 0.0

plt.plot(x, y1, 'r')
plt.plot(x, y2, 'g')
plt.plot(x, y3, 'b')
```

```python
x = np.linspace(-100.0, 100.0, 1001) 

y = x * x - 4.0 * x + 4.0

plt.plot(x, y, 'g')
```

```python
x = np.linspace(0.0, 100.0, 11) 
y1 = np.exp(x)
y2 = 1.0 / np.exp(x)
y3 = x**x

plt.plot(x, y1, 'r')
plt.plot(x, y2, 'g')
plt.plot(x, y3, 'b')
```
